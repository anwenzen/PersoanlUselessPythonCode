```python
# 7 - 9
import calendar
# import re

n = int(input())
m = [int(i) for i in input().split()]
k = 2002
ans = []
for i in range(n):
    cal = calendar.month(k+m[i], 9)
    for j in cal.split('\n'):
        if '29' in j:
            sss = j.split()
            for o in range(len(sss)):
                if sss[o]=='29':
                    ans.append(o)
week = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

for i in ans:
    print(week[i])
```

```python
# 7-13
n ,m = map(int,input().split())
a = 0
b = 0

for i in range(max(m,n)):
    if a+b>=n*m:
        break
    if i%2 == 0:
        a = a + (n+m)-(2*i+1)
    else:
        b = b + (n+m)-(2*i+1)
if a>b:
    print('trxnb')
else:
    print('trxnb')
```

```python
# 7-11
n = int(input())
for i in range(2**n):
    for j in range(i+1):
        if (i&j)==j: 
            # 按位与运算 当二进制同时为1时，该位为1，否则为0 
            # 如100101&101100 --> 100100
            print('*',end='')
        else:
            print(' ',end='')
    print()
```
