```python
# 1018 time out
n = int(input())
A = {'B':0,'C':0,'J':0}
B = {'B':0,'C':0,'J':0}
p = 0
for i in range(n):
    s = input().split()
    if s[0] =='C' and s[1]=='J':
        A['C']+=1
    if s[0] =='B' and s[1]=='C':
        A['B']+=1
    if s[0] =='J' and s[1]=='B':
        A['J']+=1
    if s[0] =='C' and s[1]=='B':
        B['B']+=1
    if s[0] =='B' and s[1]=='J':
        B['J']+=1
    if s[0] =='J' and s[1]=='C':
        B['C']+=1
    if s[0] == s[1]:
        p+=1
print(A['B']+A['J']+A['C'],p,n-p-(A['B']+A['J']+A['C']))
print(B['B']+B['J']+B['C'],p,n-p-(B['B']+B['J']+B['C'])) 
def maxW(d):
    if d["B"]>=d["C"] and d['B']>=d["J"]:
        return 'B'
    elif d["B"]<d["C"] and d['C']>=d["J"]:
        return 'C'
    else:
        return 'J'
print(maxW(A),maxW(B))
```

```python
# 1020
N, D = map(int,input().split())        # 种类数、需求量
k = [float(i) for i in input().split()]  # 库存
n = [float(i) for i in input().split()]  # 价格
l = []
for i in range(N):
    l.append({"money":n[i]/k[i],"stock":k[i]})
s = sorted(l, key=lambda i: i['money'], reverse=True)
sum = 0
for i in range(D):
    for j in range(N):
        if s[j]['stock']>=1:
            s[j]['stock'] -= 1
            sum += s[j]['money']
            break
        if 0<s[j]['stock']<1:
            sum = sum + s[j]['money']*s[j]['stock']
            if j+1<N:
                if s[j+1]['stock'] >=(1-s[j]['stock']):
                    sum = sum + s[j+1]['money'] * (1 - s[j]['stock'])
                    s[j+1]['stock'] = s[j+1]['stock'] (1 - s[j]['stock'])
            s[j]['stock']=0
            break
print(f'{sum:.2f}')
```
